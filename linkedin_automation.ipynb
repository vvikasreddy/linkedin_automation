```python
from selenium import webdriver
# from selenium.webdriver.firefox.options import Options
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
import random
```

```python
class LinkedInConnector:
    def __init__(self):
        options = Options()
        # code : "C:\Program Files\Google\Chrome\Application\chrome.exe" --remote-debugging-port=9222
        options.debugger_address = "127.0.0.1:9222"  # Matches the port from CMD
        self.driver = webdriver.Chrome(options=options)  # Update path
        
        # self.driver.get('https://www.linkedin.com/login')
        self.driver.get("https://www.linkedin.com/search/results/people") 

    def login(self):
        
        time.sleep(random.uniform(1, 3))

        email_field = WebDriverWait(self.driver, 10).until(
            EC.element_to_be_clickable((By.ID, 'username'))
        )
        password_field = WebDriverWait(self.driver, 10).until(
            EC.element_to_be_clickable((By.ID, 'password'))
        )
        
        for char in self.email:
            email_field.send_keys(char)
            time.sleep(random.uniform(0.05, 0.15))
        time.sleep(random.uniform(0.5, 1))
        
        for char in self.password:
            password_field.send_keys(char)
            time.sleep(random.uniform(0.05, 0.15))
        time.sleep(random.uniform(0.5, 1))
        
        password_field.send_keys(Keys.RETURN)
        time.sleep(random.uniform(2, 4))

    def search_company(self, company_name):
        print("Navigating to search")
        self.driver.execute_script("window.scrollBy(0, 200);")
        time.sleep(random.uniform(0.5, 1.5))
        
        search_field = WebDriverWait(self.driver, random.uniform(10, 15)).until(
            EC.element_to_be_clickable((By.XPATH, "//input[@role='combobox' and @aria-label='Search']"))
        )
        print("Search field located")
        
        self.driver.execute_script("arguments[0].scrollIntoView({behavior: 'smooth', block: 'center'});", search_field)
        time.sleep(random.uniform(0.5, 1))
        
        if not search_field.is_displayed() or not search_field.is_enabled():
            print("Search field not immediately interactable, focusing")
            self.driver.execute_script("arguments[0].focus();", search_field)
            time.sleep(random.uniform(0.3, 0.7))
        
        try:
            search_field.click()
            print("Search field clicked successfully")
        except Exception as e:
            print(f"Normal click failed: {e}, attempting JavaScript workaround")
            self.driver.execute_script("arguments[0].dispatchEvent(new Event('click', {bubbles: true}));", search_field)
            time.sleep(random.uniform(0.3, 0.7))
        
        for char in company_name:
            search_field.send_keys(char)
            time.sleep(random.uniform(0.05, 0.1))
        time.sleep(random.uniform(0.5, 1))
        
        search_field.send_keys(Keys.RETURN)
        print("Search submitted")
        time.sleep(random.uniform(1.5, 3))

    def search_school(self, company_name):
        print("Navigating to search")
        self.driver.execute_script("window.scrollBy(0, 200);")
        time.sleep(random.uniform(0.5, 1.5))
        
        search_field = WebDriverWait(self.driver, random.uniform(10, 15)).until(
            EC.element_to_be_clickable((By.XPATH, "//input[@role='combobox' and @aria-label='Search']"))
        )
        print("Search field located")
        
        self.driver.execute_script("arguments[0].scrollIntoView({behavior: 'smooth', block: 'center'});", search_field)
        time.sleep(random.uniform(0.5, 1))
        
        if not search_field.is_displayed() or not search_field.is_enabled():
            print("Search field not immediately interactable, focusing")
            self.driver.execute_script("arguments[0].focus();", search_field)
            time.sleep(random.uniform(0.3, 0.7))
        
        try:
            search_field.click()
            print("Search field clicked successfully")
        except Exception as e:
            print(f"Normal click failed: {e}, attempting JavaScript workaround")
            self.driver.execute_script("arguments[0].dispatchEvent(new Event('click', {bubbles: true}));", search_field)
            time.sleep(random.uniform(0.3, 0.7))
        
        for char in company_name:
            search_field.send_keys(char)
            time.sleep(random.uniform(0.05, 0.1))
        time.sleep(random.uniform(0.5, 1))
        
        search_field.send_keys(Keys.RETURN)
        print("Search submitted")
        time.sleep(random.uniform(1.5, 3))

    def add_company(self, company_name="Oracle"):
        try:
            # Step 1: Click the "Add a company" button
            add_company_button = WebDriverWait(self.driver, 10).until(
                EC.element_to_be_clickable((By.XPATH, "//button[contains(@class, 'reusable-search-filters-advanced-filters__add-filter-button') and .//span[text()='Add a company']]"))
            )
            add_company_button.click()
        except:
            print("Add company might have already been clicked")

        time.sleep(random.uniform(0.5, 1.5))  # Wait for input to appear

        # Step 2: Find the company input field and add company name
        if company_name:
            company_input = WebDriverWait(self.driver, 10).until(
                EC.element_to_be_clickable((By.XPATH, "//input[@placeholder='Add a company']"))
            )
            print("Company input located")
            company_input.send_keys(company_name)

            # Step 3: Select the first suggestion
            time.sleep(random.uniform(0.5, 1.5))  # Wait for suggestions to load
            company_input.send_keys(Keys.DOWN)    # Press Down Arrow to select first suggestion
            time.sleep(random.uniform(0.5, 1.5))  # Brief pause for selection
            company_input.send_keys(Keys.ENTER)   # Press Enter to confirm
            print("Company selected")

    def add_location(self, location="Baltimore, MD"):
        try:
            # Step 1: Click the "Add a location" button
            add_location_button = WebDriverWait(self.driver, 10).until(
                EC.element_to_be_clickable((By.XPATH, "//button[contains(@class, 'reusable-search-filters-advanced-filters__add-filter-button') and .//span[text()='Add a location']]"))
            )
            add_location_button.click()
        except:
            print("Add location might have already been clicked")

        time.sleep(random.uniform(0.5, 1.5))  # Wait for input to appear

        # Step 2: Find the location input field and add location
        if location:
            location_input = WebDriverWait(self.driver, 10).until(
                EC.element_to_be_clickable((By.XPATH, "//input[@placeholder='Add a location']"))
            )
            print("Location input located")
            location_input.send_keys(location)

            # Step 3: Select the first suggestion
            time.sleep(random.uniform(0.5, 1.5))  # Wait for suggestions to load
            location_input.send_keys(Keys.DOWN)    # Press Down Arrow to select first suggestion
            time.sleep(random.uniform(0.5, 1.5))  # Brief pause for selection
            location_input.send_keys(Keys.ENTER)   # Press Enter to confirm
            print("Location selected")

    def apply_filters(self, university=False, role=None, company = None):
        
        print("Applying filters")
        # Step 1: Open the "All filters" modal
        all_filters_button = WebDriverWait(self.driver, 10).until(
            EC.element_to_be_clickable((By.XPATH, "//button[contains(@class, 'search-reusables__all-filters-pill-button') and text()='All filters']"))
        )
        all_filters_button.click()
        time.sleep(random.uniform(0.5, 1.5))  # Wait for modal to load
        # step 1: Find the scrollable container
        container = self.driver.find_element(By.CLASS_NAME, "artdeco-modal__content")
        
        self.add_location("United States")
        self.driver.execute_script("arguments[0].scrollTop += 500;", container)
        self.add_company(company)

        # stepp 2: Scroll the container down by 500 pixels
        self.driver.execute_script("arguments[0].scrollTop += 500;", container)

        time.sleep(random.uniform(0.5, 1.5))  # Wait for content to load

        try: 
            # Step 3: Click the "Add a school" button
            add_school_button = WebDriverWait(self.driver, 10).until(
                EC.element_to_be_clickable((By.XPATH, "//button[contains(@class, 'reusable-search-filters-advanced-filters__add-filter-button') and .//span[text()='Add a school']]"))
            )
            add_school_button.click()
        except:
            print("Add school might have already clicked")

            
        time.sleep(random.uniform(0.5, 1.5))  # Wait for input to appear

        # step 4: Find the school input field
        if university: 
            school_input = WebDriverWait(self.driver, 10).until(
                EC.element_to_be_clickable((By.XPATH, "//input[@placeholder='Add a school']"))
            )
            print("School input located")
            school_input.send_keys("University of Maryland Baltimore County")

            # step 5: Select the first suggestion
            time.sleep(random.uniform(0.5, 1.5))  # Wait for suggestions to load
            school_input.send_keys(Keys.DOWN)    # Press Down Arrow to select first suggestion
            time.sleep(random.uniform(0.5, 1.5))  # Brief pause for selection
            school_input.send_keys(Keys.ENTER)   # Press Enter to confirm
            print("School selected")
        
        
    

        #setp 6: Click the show results button
        show_results_button = WebDriverWait(self.driver, 10).until(
            EC.element_to_be_clickable((By.XPATH, "//button[.//span[text()='Show results']]"))
        )
        print("Show results button located")
        show_results_button.click()
    
    def connect_with_people(self, prev_names, connector_msg, connection_msg):
        count = 0
        span = 1
        driver = self.driver
        wait = WebDriverWait(driver, 10)
        # driver = self.driver
        print(prev_names)
        names = []

        while count <= 20:
           
            
            print('trying to find the ul element')
            # Find the first <ul> element on the page (or adjust to target a specific one)
            ul_element = wait.until(
                EC.element_to_be_clickable((By.XPATH, "//ul[@role='list']"))  # Stronger condition
            )
            ul_class = ul_element.get_attribute("class")  # Should work if fetched recently
            print(f"Class name of the <ul> element: {ul_class}")

            # List to store names
            

            li_elements = ul_element.find_elements(By.TAG_NAME, "li")
            print(li_elements[0].get_attribute("class"))

            # Function to simulate human typing
            def human_typing(element, text):
                for char in text:
                    element.send_keys(char)
                    time.sleep(random.uniform(0.05, 0.15))  # Random delay between 50ms and 150ms per character

            def enter_message_text(driver, message):

                print("find the text box")
                # Step 1: Find and click the textbox
                textbox = driver.find_element(
                    By.XPATH, "//div[@contenteditable='true'][@aria-label='Write a message…']"
                )
                time.sleep(random.uniform(1.0, 2.0))  # Human-like pause before clicking
                textbox.click()

                print("send text to the textbox")
                # Step 2: Send text to the textbox
                # message = "Hello, this is a test message!"
                human_typing(textbox, message)
                time.sleep(random.uniform(1.0, 2.0))  # Pause after typing

                # remove this while actual implementation------------------------------: Clear the text
                # print("Falling back to manual clearing")
                # textbox.send_keys(Keys.CONTROL + "a")  # Select all
                # time.sleep(random.uniform(0.5, 1.0))
                # textbox.send_keys(Keys.BACKSPACE)  # Delete
                # print("Textbox cleared manually")

                print("# Step 3: Find and click the Send button")
                send_button = driver.find_element(
                    By.XPATH, "//button[@type='submit'][contains(@class, 'msg-form__send-button')]"
                )
                time.sleep(random.uniform(1.0, 2.0))  # Pause before clicking
# ----------------------------Uncomment to actually send the message--------------------------------
                send_button.click()

                # Step 4: 
                close_button = WebDriverWait(driver, 10).until(
                    EC.presence_of_element_located((By.XPATH, "//span[contains(., 'Close your conversation')]/parent::button"))
                )
                # print(f"Found button with ID: {button.get_attribute('id')}")  # Should be "ember366"
                close_button.click()

                # step 5:
                # Find the Send button
                # send_button = driver.find_element(
                #     By.XPATH, "//button[@class='msg-form__send-button artdeco-button artdeco-button--1' and @type='submit']"
                # )
                # time.sleep(random.uniform(1.0, 2.0))  # Human-like pause before clicking

                # # Click the button
                # send_button.click()
                # print("Send button clicked successfully!")
            
                time.sleep(random.uniform(1.0, 2.0))
                time.sleep(random.uniform(1.0, 2.0))  # Pause after submitting
                # Iterate through each <li> and find the "Connect" button
            for li in li_elements:
                try:
                    # Random initial delay to simulate human hesitation
                    time.sleep(random.uniform(1.5, 3.0))
                    # print(li.get_attribute("class"))


                    # Find the "Connect" button within this <li>
                    connect_button = li.find_element(
                        By.XPATH, ".//button[contains(@class, 'artdeco-button')][.//span[text()='Connect']]"
                    )
                    
                    if connect_button.get_attribute("aria-label") in prev_names:
                        print("Already connected")
                        continue

                    # Store the name from aria-label
                    names.append(connect_button.get_attribute("aria-label"))
                    
                    # Simulate human pause before clicking
                    # time.sleep(random.uniform(0.8, 2.0))
                    connect_button.click()
                    
                    time.sleep(random.uniform(1.0, 3.0))
                    # Wait for "Add a note" button with a random pause afterward
                    add_note_button = wait.until(
                        EC.element_to_be_clickable(
                            (By.XPATH, "//button[@aria-label='Add a note'][contains(@class, 'artdeco-button')]")
                        )
                    )
                    time.sleep(random.uniform(0.5, 1.5))
                    add_note_button.click()

                    # Wait for the textarea to appear
                    textarea = wait.until(
                        EC.visibility_of_element_located(
                            (By.XPATH, "//textarea[@id='custom-message'][contains(@class, 'connect-button-send-invite__custom-message')]")
                        )
                    )

                    # # Slightly vary the message to avoid repetition detection
                    # messages = [
                    #     "Hi, we met at the recent tech conference. Looking forward to connecting!",
                    #     "Hey there, great meeting you at the tech conference! Excited to connect.",
                    #     "Hi! We crossed paths at the tech conference – happy to connect here!"
                    # ]
                    
                    message = f"Hey {names[-1].split(' ')[1]}, " + connector_msg

                    # Simulate human typing with random delays
                    time.sleep(random.uniform(0.3, 1.0))  # Pause before typing
                    human_typing(textarea, message)

                    # Wait before clicking "Send"
                    send_button = wait.until(
                        EC.element_to_be_clickable(
                            (By.XPATH, "//button[@aria-label='Send invitation'][contains(@class, 'artdeco-button')]")
                        )
                    )

                    # Print confirmation (optional)
                    print("Found 'Send' button with aria-label: ", send_button.get_attribute("aria-label"))

                    # Random pause before clicking send
                    time.sleep(random.uniform(1.0, 2.5))
# ----------------------------Uncomment to actually send the message--------------------------------
                    # Uncomment to actually send
                    send_button.click()

                    # For testing: Cancel the send with a human-like delay
                    # cancel_send = wait.until(
                    #     EC.element_to_be_clickable(
                    #         (By.XPATH, "//button[@aria-label='Dismiss'][contains(@class, 'artdeco-button')]")
                    #     )
                    # )
                    count += 1
                    time.sleep(random.uniform(0.7, 1.8))
                    # cancel_send.click()

                    # Add a longer random delay between iterations to mimic natural browsing
                    time.sleep(random.uniform(3.0, 6.0))

                except Exception as e:
                    print(f"Exception occurred: --------")
                    # Fallback to "Message" button if "Connect" fails
                    try:
                        connect_button = li.find_element(
                            By.XPATH, ".//button[contains(@class, 'artdeco-button')][.//span[text()='Message']]"
                        )
                        # Store the name from aria-label
                        names.append(connect_button.get_attribute("aria-label"))
                        time.sleep(random.uniform(1.0, 2.0))  # Human-like pause before moving on
                        connect_button.click()
                        time.sleep(random.uniform(1.0, 2.0))  # Another pause before moving on
                        message = f"Hey {names[-1].split(' ')[1]}, " + connection_msg
                        try:
                            print("checkimg for the button :")
                            buttons = driver.find_elements(By.XPATH, "//button[@aria-label='Create video meeting']")
                            if buttons:
                                print("Button with aria-label='Create video meeting' is present.")
                            else:
                                print("Button with aria-label='Create video meeting' is not present.")
                                raise Exception("Button not found")
                            time.sleep(random.uniform(1.0, 2.0))
                            
                            print("Button with aria-label='Create video meeting' exists.")
                            enter_message_text(driver, message)
                            count += 1

                        except Exception as e:
                            print("Button with aria-label='Create video meeting' does not exist.")
                            close_button = WebDriverWait(driver, 10).until(
                                EC.element_to_be_clickable(
                                    (By.XPATH, "//button[.//span[text()='Close your draft conversation']]")
                                )
                            )
                            time.sleep(random.uniform(1.0, 2.0))
                            close_button.click()
                            continue   

                    # except NoSuchElementException:
                    #     print("Button with aria-label='Create video meeting' does not exist.")
                        
                    #     # enter_message_text(driver)
                    #     continue

                    except Exception as e:
                        # if e == "NoSuchElementException":
                        print("No 'Message' button found either. Skipping...")
                    continue  # Move to the next <li> gracefully

            # Optional: Add a final delay to simulate a human finishing up
            time.sleep(random.uniform(2.0, 5.0))
            print(f"Processed {len(names)} connection requests.")
            span += 1
            driver.execute_script("window.scrollBy(0, 200);") 
            text = "Page " + str(span)
            page_button = WebDriverWait(driver, 10).until(
                EC.element_to_be_clickable((By.XPATH, f"//button[@aria-label='{text}']"))
            )
            time.sleep(random.uniform(1.0, 2.0))  # Human-like pause before clicking
            page_button.click()
        prev_names = prev_names + names
        return prev_names
```

```python
"""
decode_google_doc.py

Download a Google Doc that lists Unicode characters with their X‑Y coordinates
and print the resulting 2‑D grid so the hidden message appears.

Expected document line format (one item per line, any amount of whitespace):
    <char-or-codepoint> <x> <y>

The first token may be either
    * the literal character  (e.g. "█"), or
    * a code‑point like "U+2588" or "0x2588".

Coordinates are non‑negative integers.  
(0, 0) is the **upper‑left** corner, X grows →, Y grows ↓.

Example line:
    U+2588  0  0
"""

from __future__ import annotations

import re
import sys
from typing import Dict, Tuple

import requests


def decode_google_doc(url: str) -> None:
    """
    Fetch *url* (a Google‑Docs link), parse the coordinate table,
    and print the reconstructed character grid.

    A Google Doc can be downloaded as plain‑text via:
        https://docs.google.com/document/d/<DOC_ID>/export?format=txt
    so *url* can be either the “normal” viewing link or the /export link.

    Parameters
    ----------
    url : str
        Link to the Google Document that contains the data.
    """

    text = _download_doc(url)
    mapping, max_x, max_y = _parse_lines(text.splitlines())

    # Build and print the grid
    for y in range(max_y + 1):
        row_chars = [
            mapping.get((x, y), " " if _is_nbsp_needed(mapping) else " ")  # regular space by default
            for x in range(max_x + 1)
        ]
        print("".join(row_chars))


# --------------------------------------------------------------------------- #
# Helpers
# --------------------------------------------------------------------------- #

_RE_CODEPOINT = re.compile(r"^(?:U\+|0x)([0-9A-Fa-f]+)$")


def _download_doc(url: str) -> str:
    """
    Download the document as plain text.  If *url* is a regular view link,
    convert it to the /export?format=txt endpoint.
    """
    if "/export" not in url:
        # Convert “…/d/<id>/edit…” to “…/d/<id>/export?format=txt”
        match = re.search(r"/document/d/([^/]+)/", url)
        if not match:
            raise ValueError("Unable to recognise Google‑Doc ID in URL.")
        doc_id = match.group(1)
        url = f"https://docs.google.com/document/d/{doc_id}/export?format=txt"

    response = requests.get(url, timeout=30)
    response.raise_for_status()
    return response.text


def _parse_lines(
    lines: list[str],
) -> Tuple[Dict[Tuple[int, int], str], int, int]:
    """
    Parse <char> <x> <y> lines into a {(x, y): char} map and track max_x, max_y.
    """
    mapping: dict[tuple[int, int], str] = {}
    max_x = max_y = 0

    for raw in lines:
        line = raw.strip()
        if not line:
            continue

        parts = line.split()
        if len(parts) != 3:
            continue  # skip malformed / header lines

        token, x_str, y_str = parts
        char = _token_to_char(token)
        try:
            x = int(x_str)
            y = int(y_str)
        except ValueError:
            continue  # skip malformed coordinates

        mapping[(x, y)] = char
        max_x = max(max_x, x)
        max_y = max(max_y, y)

    if not mapping:
        raise ValueError("No character data found in document.")
    return mapping, max_x, max_y


def _token_to_char(token: str) -> str:
    """
    Convert first field into a single Unicode character.
    Accepts direct char or “U+XXXX / 0xXXXX” hexadecimal codepoints.
    """
    m = _RE_CODEPOINT.match(token)
    if m:
        codepoint = int(m.group(1), 16)
        return chr(codepoint)
    return token[0]  # take the first character literally


def _is_nbsp_needed(mapping: Dict[Tuple[int, int], str]) -> bool:
    """
    Heuristic: if the grid contains non‑ASCII blocks like '█', printing with
    a normal space sometimes collapses visually in some terminals.  Using an
    NBSP (U+00A0) keeps monospacing reliable.
    """
    for ch in mapping.values():
        if ord(ch) > 0x7F:
            return True
    return False


# --------------------------------------------------------------------------- #
# CLI entry‑point for quick testing
# --------------------------------------------------------------------------- #

if __name__ == "__main__":
    if len(sys.argv) != 2:
        print("Usage: python decode_google_doc.py https://docs.google.com/document/d/e/2PACX-1vQGUck9HIFCyezsrBSnmENk5ieJuYwpt7YHYEzeNJkIb9OSDdx-ov2nRNReKQyey-cwJOoEKUhLmN9z/pub")
        sys.exit(1)
    decode_google_doc(sys.argv[1])
```
