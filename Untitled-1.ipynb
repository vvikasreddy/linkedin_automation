```python
from qdrant_client import QdrantClient
from qdrant_client.http.models import VectorParams, Distance, PointStruct
import json
```

```python
# 2. Load BGE (small English) from Hugging Face
from transformers import AutoTokenizer, AutoModel
MODEL_NAME = "BAAI/bge-large-en"   # ⟵ 384-dim; switch to "-large-en" for 1024-dim
tokenizer = AutoTokenizer.from_pretrained(MODEL_NAME)
model     = AutoModel.from_pretrained(MODEL_NAME).eval()

```

```python
import torch
device = "cuda" if torch.cuda.is_available() else "cpu"
model.to(device)
```

```python
# 2. Connect & recreate collection
client = QdrantClient(url="http://localhost:6333")
client.recreate_collection(
    collection_name="messages",
    vectors_config=VectorParams(size=1024, distance=Distance.COSINE),
)
```

```python
def embed_text(text: str) -> list[float]:
    # basic BGE pooling: CLS token + L2 norm
    inputs = tokenizer(text, return_tensors="pt", truncation=True, padding=True)
    inputs = {k: v.to(device) for k, v in inputs.items()}
    with torch.no_grad():
        out = model(**inputs, return_dict=True)
    vec = out.last_hidden_state[:, 0]                # [batch, dim]
    vec = vec / vec.norm(p=2, dim=-1, keepdim=True)  # normalize
    return vec[0].cpu().tolist()
```

```python
# 4. Load your JSON file
with open("messages_data4.json", "r", encoding="utf-8") as f:
    records = json.load(f)   # expect a list of dicts
```

```python
# 5. Embed & batch-upsert
batch_size = 64
points: list[PointStruct] = []
for idx, rec in enumerate(records):
    emb = embed_text(rec["message_text"])
    points.append(PointStruct(id=idx, vector=emb, payload=rec))
    # flush in batches
    if len(points) >= batch_size:
        client.upsert(collection_name="messages", points=points)
        points.clear()
```

```python

# any leftovers
if points:
    client.upsert(collection_name="messages", points=points)

print(f"✅ Imported {len(records)} messages into Qdrant ‘messages’ collection.")
```

```python
import json
from sentence_transformers import SentenceTransformer
from qdrant_client import QdrantClient
from qdrant_client.http.models import PointStruct

# 1️⃣ Setup your embedder & Qdrant client
embedder = SentenceTransformer("sentence-transformers/all-mpnet-base-v2", device="cuda")
client   = QdrantClient(url="http://localhost:6333")

# 2️⃣ (Re)create collection once
from qdrant_client.http.models import VectorParams, Distance
client.recreate_collection(
    collection_name="messages",
    vectors_config=VectorParams(size=768, distance=Distance.COSINE),
)
```

```python

# 3️⃣ Define your files and batch size
files = [f"messages_data{i}.json" for i in range(1,5)]
batch_size = 32
buffer = []
id_counter = 0

def record_to_text(rec: dict) -> str:
    # Option A: dump the whole dict as JSON
    return json.dumps(rec, ensure_ascii=False)
```

```python

# 4️⃣ Loop through each file
for fname in files:
    with open(fname, "r", encoding="utf-8") as f:
        records = json.load(f)   # list of dicts
    for rec in records:
        # embed
        full_record_text = record_to_text(rec)
        
        emb = embedder.encode(full_record_text, convert_to_numpy=True).tolist()
        # prepare point
        buffer.append(PointStruct(id=id_counter, vector=emb, payload=rec))
        id_counter += 1

        # flush batch
        if len(buffer) >= batch_size:
            client.upsert(collection_name="messages", points=buffer)
            buffer.clear()
    

# 5️⃣ Flush any remaining points
if buffer:
    client.upsert(collection_name="messages", points=buffer)

print(f"✅ Imported {id_counter} total messages from {len(files)} files.")
```

```python
query_text = "Joel Kronander KLA"

# 1. Embed your question
query_embedding = embedder.encode(query_text, convert_to_numpy=True).tolist()

# 2. Perform the search
results = client.search(
    collection_name="messages",
    query_vector=query_embedding,
    limit=5,
    with_payload=True
)

# 3. Display the hits
for hit in results:
    print(f"ID: {hit.id}  Score: {hit.score:.4f}")
    print(f"Date: {hit.payload['message_date']}")
    print(f"Text: {hit.payload['message_text']}")
    print(f"Sender: {hit.payload['message_id']}")
    print("-" * 40)
```
